```python
import os

# Define the path to the desired folder in your Google Drive
# This path is relative to the mount point '/content/drive/'
# Example: 'My Drive/YouTube_MP3s' will create/use a folder named 'YouTube_MP3s' in your main Google Drive
drive_upload_folder = 'YouTube_MP3s'
drive_upload_folder_path = f'/content/drive/My Drive/{drive_upload_folder}'

# Create the directory in Google Drive if it doesn't exist
os.makedirs(drive_upload_folder_path, exist_ok=True)
print(f"Google Drive upload folder set to: {drive_upload_folder_path}")

# --- User Input ---
# Paste the YouTube video or playlist URL here
youtube_url = 'https://youtu.be/ztq7-kkygZk?feature=shared'  # <--- REPLACE WITH YOUR URL
# --- End User Input ---

print(f"Input URL set to: {youtube_url}")
```

```python
import yt_dlp
import os

# Define a temporary directory for downloads
temp_download_dir = '/content/temp_downloads'
os.makedirs(temp_download_dir, exist_ok=True)

# yt-dlp options for downloading and converting to MP3
ydl_opts = {
    'format': 'bestaudio/best', # Download the best quality audio format
    'postprocessors': [{
        'key': 'FFmpegExtractAudio', # Use FFmpeg to extract audio
        'preferredcodec': 'mp3',    # Convert to MP3
        'preferredquality': '0',   # Highest quality MP3
    }],
    'outtmpl': os.path.join(temp_download_dir, '%(title)s.%(ext)s'), # Output filename template in temp dir
    'ignoreerrors': True, # Continue if some videos in a playlist fail
}

print(f"Downloading and converting: {youtube_url}")

try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(youtube_url, download=True)
        # info_dict contains metadata about the downloaded video(s)/playlist
    print("Download and conversion complete.")

except Exception as e:
    print(f"An error occurred during download/conversion: {e}")
```

```python
import shutil
import glob
import os

print(f"Moving MP3 files to Google Drive folder: {drive_upload_folder_path}")

# Find all .mp3 files in the temporary download directory
mp3_files = glob.glob(os.path.join(temp_download_dir, '*.mp3'))

if not mp3_files:
    print("No MP3 files found to upload.")
else:
    for mp3_file in mp3_files:
        file_name = os.path.basename(mp3_file)
        destination_path = os.path.join(drive_upload_folder_path, file_name)

        try:
            shutil.move(mp3_file, destination_path)
            print(f"Uploaded '{file_name}' to Google Drive.")
        except Exception as e:
            print(f"Error uploading '{file_name}': {e}")

    print("All available MP3 files processed for upload.")

# Optional: Clean up temporary directory
# shutil.rmtree(temp_download_dir)
# print("Temporary download directory cleaned up.")
```
