# YouTube to 2-Stem Reggae Splitter
This notebook downloads a reggae track from YouTube and splits it into vocals and accompaniment using Spleeter (2 stems).

```python
# Install dependencies
!pip install yt-dlp spleeter
```

```python
# Download a reggae track
!yt-dlp -x --audio-format mp3 --output "test_reggae.mp3" https://www.youtube.com/watch?v=2sUuLfaI7xk
```

```python
# Split into 2 stems (vocals + accompaniment)
!spleeter separate -p spleeter:2stems -o output test_reggae.mp3
```
