```python
import os
from demucs.pretrained import get_model
from demucs.apply import apply_model
import torchaudio
import torch

# Set input/output paths
input_audio = "test_reggae.mp3"
output_dir = "demucs_output"

# Load audio
waveform, sr = torchaudio.load(input_audio)
waveform = waveform.mean(dim=0, keepdim=True)  # convert to mono if needed
waveform = torchaudio.functional.resample(waveform, sr, 44100)
sr = 44100

# Load Demucs model
model = get_model(name='htdemucs')  # Use 'htdemucs' for better quality
model.eval()

# Separate stems
with torch.no_grad():
    sources = apply_model(model, waveform, sr=sr, device='cpu')[0]

# Save output stems
os.makedirs(output_dir, exist_ok=True)
for name, source in zip(model.sources, sources):
    path = os.path.join(output_dir, f"{name}.wav")
    torchaudio.save(path, source.cpu(), sr)
    print(f"Saved {path}")
```
