```python
# Install dependencies if not already installed
!pip install -q pytube pydub mutagen

from pytube import YouTube
from pydub import AudioSegment
from mutagen.easyid3 import EasyID3
from mutagen.mp3 import MP3
import os

# Ensure output folder exists
os.makedirs(drive_upload_folder_path, exist_ok=True)

for idx, url in enumerate(youtube_urls, 1):
    try:
        print(f"\n[{idx}/{len(youtube_urls)}] Processing: {url}")
        yt = YouTube(url)
        stream = yt.streams.filter(only_audio=True).first()
        temp_filename = f"temp_audio_{idx}"
        out_file = stream.download(filename=temp_filename)

        # Convert and trim
        audio = AudioSegment.from_file(out_file)
        trimmed = audio[:clip_length_slider.value * 1000]
        final_filename = f"{yt.title[:50].strip().replace(' ', '_')}.mp3"
        final_path = os.path.join(drive_upload_folder_path, final_filename)
        trimmed.export(final_path, format="mp3")
        os.remove(out_file)

        # Tag the MP3
        audiofile = MP3(final_path, ID3=EasyID3)
        audiofile['title'] = yt.title
        audiofile['artist'] = yt.author
        audiofile['album'] = 'YouTube Batch'
        audiofile.save()

        print(f"✅ Saved and tagged: {final_filename}\n→ {final_path}")

    except Exception as e:
        print(f"❌ Error processing {url}: {e}")
```

```python
import yt_dlp
import os

# Define a temporary directory for downloads
temp_download_dir = '/content/temp_downloads'
os.makedirs(temp_download_dir, exist_ok=True)

# yt-dlp options for downloading and converting to MP3
ydl_opts = {
    'format': 'bestaudio/best', # Download the best quality audio format
    'postprocessors': [{
        'key': 'FFmpegExtractAudio', # Use FFmpeg to extract audio
        'preferredcodec': 'mp3',    # Convert to MP3
        'preferredquality': '0',   # Highest quality MP3
    }],
    'outtmpl': os.path.join(temp_download_dir, '%(title)s.%(ext)s'), # Output filename template in temp dir
    'ignoreerrors': True, # Continue if some videos in a playlist fail
}

print(f"Downloading and converting: {youtube_url}")

try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(youtube_url, download=True)
        # info_dict contains metadata about the downloaded video(s)/playlist
    print("Download and conversion complete.")

except Exception as e:
    print(f"An error occurred during download/conversion: {e}")
```
