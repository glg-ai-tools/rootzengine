# YouTube to MP3 Converter
This notebook allows you to download audio from a YouTube link and convert it to an MP3 file.
You can choose the desired length (in seconds) of the MP3 output.

```python
from pytube import YouTube
from pydub import AudioSegment
from IPython.display import display
import ipywidgets as widgets
import os

# Dropdown menu for selecting length
length_options = [30, 60, 90, 120, 180, 300]  # seconds
length_dropdown = widgets.Dropdown(
    options=length_options,
    value=60,
    description='MP3 Length:',
    style={'description_width': 'initial'}
)
display(length_dropdown)
```

```python
# Input for YouTube link
url_input = widgets.Text(
    value='',
    placeholder='Enter YouTube URL here',
    description='YouTube URL:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='70%')
)
display(url_input)
```

```python
# Conversion function
def convert_youtube_to_mp3(url, clip_length):
    yt = YouTube(url)
    stream = yt.streams.filter(only_audio=True).first()
    out_file = stream.download(filename='temp_audio')
    audio = AudioSegment.from_file(out_file)
    trimmed = audio[:clip_length * 1000]
    output_filename = f"{yt.title[:50].strip().replace(' ', '_')}.mp3"
    trimmed.export(output_filename, format="mp3")
    os.remove(out_file)
    print(f"Exported: {output_filename}")
    return output_filename
```

```python
# Run conversion
if url_input.value:
    result = convert_youtube_to_mp3(url_input.value, length_dropdown.value)
else:
    print("Please enter a valid YouTube URL.")
```
